## Theory Questions and Answers

**Q: What is a random variable in probability theory**

**A:** A random variable is a variable whose possible values are numerical outcomes of a random phenomenon.

**Q: What are the types of random variables**

**A:** There are two types: discrete (countable outcomes) and continuous (infinite possible values within a range).

**Q: What is the difference between discrete and continuous distributions**

**A:** Discrete distributions involve countable outcomes; continuous distributions involve an infinite range of values.

**Q: What are probability distribution functions (PDF)**

**A:** A PDF gives the probability that a discrete random variable is equal to a specific value.

**Q: How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)**

**A:** CDF gives the probability that a random variable is less than or equal to a certain value, summing up PDF values.

**Q: What is a discrete uniform distribution**

**A:** It's a distribution where all outcomes are equally likely and finite.

**Q: What are the key properties of a Bernoulli distribution**

**A:** Two outcomes (0 and 1), probability p for success, 1-p for failure.

**Q: What is the binomial distribution, and how is it used in probability**

**A:** It models number of successes in a fixed number of independent Bernoulli trials.

**Q: What is the Poisson distribution and where is it applied**

**A:** It models the number of events occurring in a fixed interval of time/space.

**Q: What is a continuous uniform distribution**

**A:** A distribution where every value within a range is equally likely.

**Q: What are the characteristics of a normal distribution**

**A:** Symmetrical, bell-shaped, mean=median=mode, defined by mean and standard deviation.

**Q: What is the standard normal distribution, and why is it important**

**A:** Normal distribution with mean=0 and std=1; used in Z-tests and statistical inference.

**Q: What is the Central Limit Theorem (CLT), and why is it critical in statistics**

**A:** It states that the sampling distribution of the mean approaches a normal distribution as the sample size increases.

**Q: How does the Central Limit Theorem relate to the normal distribution**

**A:** It justifies using the normal distribution for inference even if the population is not normal.

**Q: What is the application of Z statistics in hypothesis testing**

**A:** To determine if there is a significant difference between sample and population means.

**Q: How do you calculate a Z-score, and what does it represent**

**A:** Z = (X - mean) / std; represents how many std deviations a value is from the mean.

**Q: What are point estimates and interval estimates in statistics**

**A:** Point estimates give single value estimates; interval estimates give a range (like confidence intervals).

**Q: What is the significance of confidence intervals in statistical analysis**

**A:** They provide a range of values likely to contain the population parameter.

**Q: What is the relationship between a Z-score and a confidence interval**

**A:** Z-scores determine the bounds of confidence intervals around a mean.

**Q: How are Z-scores used to compare different distributions**

**A:** They standardize values from different distributions for comparison.

**Q: What are the assumptions for applying the Central Limit Theorem**

**A:** Independent samples, sufficiently large sample size, finite variance.

**Q: What is the concept of expected value in a probability distribution**

**A:** The long-run average value of repetitions of the experiment it represents.

**Q: How does a probability distribution relate to the expected outcome of a random variable?**

**A:** It helps calculate the expected value based on weighted probabilities of all outcomes.

## Practical Tasks (Python Code)

### 1. Generate a random variable and display its value

```python
import random
random_variable = random.randint(1, 100)
print('Random Variable:', random_variable)
```

### 2. Generate a discrete uniform distribution and plot PMF

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import randint

x = np.arange(1, 7)
pmf = [1/6]*6
plt.bar(x, pmf)
plt.title("Discrete Uniform Distribution - PMF")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.show()
```

### 3. Bernoulli PDF

```python
from scipy.stats import bernoulli
p = 0.3
x = [0, 1]
pmf = bernoulli.pmf(x, p)
print("PDF of Bernoulli distribution (p=0.3):", pmf)
```

### 4. Binomial distribution simulation and histogram

```python
from scipy.stats import binom
data = binom.rvs(n=10, p=0.5, size=1000)
plt.hist(data, bins=11, edgecolor='black')
plt.title("Binomial Distribution Histogram (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

### 5. Poisson distribution

```python
from scipy.stats import poisson
data = poisson.rvs(mu=3, size=1000)
plt.hist(data, bins=range(0, max(data)+1), edgecolor='black')
plt.title("Poisson Distribution Histogram (mu=3)")
plt.xlabel("Occurrences")
plt.ylabel("Frequency")
plt.show()
```

### 6. CDF of a discrete uniform distribution

```python
cdf = np.cumsum(pmf)
plt.step(x, cdf, where='post')
plt.title("CDF - Discrete Uniform Distribution")
plt.xlabel("Outcome")
plt.ylabel("Cumulative Probability")
plt.show()
```

### 7. Continuous uniform distribution using NumPy

```python
data = np.random.uniform(low=0, high=10, size=1000)
plt.hist(data, bins=30, edgecolor='black')
plt.title("Continuous Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

### 8. Normal distribution simulation and histogram

```python
data = np.random.normal(loc=0, scale=1, size=1000)
plt.hist(data, bins=30, edgecolor='black')
plt.title("Normal Distribution Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

### 9. Calculate Z-scores and plot

```python
from scipy.stats import zscore
z_scores = zscore(data)
plt.hist(z_scores, bins=30, edgecolor='black')
plt.title("Z-score Distribution")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()
```

### 10. Central Limit Theorem demonstration

```python
means = []
for _ in range(1000):
    sample = np.random.exponential(scale=2, size=30)
    means.append(np.mean(sample))
plt.hist(means, bins=30, edgecolor='black')
plt.title("Central Limit Theorem (Exponential Distribution Sample Means)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```
